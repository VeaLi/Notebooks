```python
!mkdir -p "data"
!mkdir -p "analysis"
```

```python
!mkdir -p "analysis/hisat2/"
!date
!hisat2 -p 4 --new-summary -x  Homo_sapiens_assembly -U "14.1.fastq"  "14.2.fastq" 2> "analysis/hisat2/hisat2.log" \
                                                                 | samtools view -b - > "analysis/hisat2/reads.raw.bam"
!date

!ls "analysis/hisat2/"
#!cat $OUTDIR/$TAG.log

##### Step 4: post-processing of alignment

# post-processing the alignments

!date
!samtools sort -@ 4 -O bam "analysis/hisat2/reads.raw.bam" > "analysis/hisat2/reads.bam" && \
  samtools index "analysis/hisat2/reads.bam" && \
  rm -v "analysis/hisat2/reads.raw.bam"
!date
 
##### Step 7: counting reads

!mkdir -p "analysis/featureCounts/"
!date
!featureCounts -a "Homo_sapiens.GRCh38.100.gtf" -s 2 -o "analysis/featureCounts/reads.fc.txt" "analysis/hisat2/reads.bam" |& tee "analysis/featureCounts/reads.log"
!date
```
